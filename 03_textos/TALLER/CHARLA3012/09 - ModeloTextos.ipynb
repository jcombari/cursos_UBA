```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("/content/drive/MyDrive/NLP Specialization/MODELO/Historia.csv")# cargar un csv a dataframe de pandas
```

```python
data.head()
```

```python
textos = data.iloc[:,1:]
textos
```

```python
X = textos.iloc[:,3:8].values
```

```python
type(X)
```

```python
X
```

```python
Y = textos.iloc[:,10].values 
```

```python
type(Y)
```

```python
Y
```

```python
xtr,xtst,ytr,ytst=train_test_split(X,Y,test_size=.2,random_state=564)
```

```python
xtrn = xtr / 10000
```

```python
xtstn = xtst / 10000
```

```python
len(xtr)
```

```python
len(ytr)
```

```python
len(xtst)
```

```python
len(ytst)
```

```python
xtstn.shape[1:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
modelo = Sequential()
```

```python
modelo.add(Dense(128,input_shape = xtstn.shape[1:]))
```

```python
modelo.add(Dense(6,activation="softmax"))
```

```python
modelo.compile(optimizer = "adam", loss = "sparse_categorical_crossentropy", metrics =["acc"] )
```

```python
modelo.fit(xtrn,ytr,epochs=120,batch_size=105,validation_split=.1)
```

```python
modelo.evaluate(xtstn,ytst)
```

```python
len(xtst)
```

```python
aciertos = 0
for i in range(len(xtst)):
     if np.argmax(modelo.predict(xtst[i].reshape(1, -1))) == ytst[i]:
       aciertos +=1
print("aciertos: ",aciertos)       
```

```python
print("Accuracy: ", aciertos / len(xtst))
```

```python
modelo.save_weights("/content/drive/MyDrive/NLP Specialization/MODELO/Historia.h5")
```
