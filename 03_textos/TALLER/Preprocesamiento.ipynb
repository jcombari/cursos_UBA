```python
import nltk                                # Python library for NLP
```

```python
# download the stopwords from NLTK
nltk.download('stopwords')
```

```python
import re
import string   
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer   
from nltk.tokenize import TweetTokenizer 
```

```python
textosSlice = textos.iloc[:100,:]
```

```python
textoPrueba = textosSlice["Texto"].values[10]
textoPrueba
```

```python
print()
print('\033[92m' + textoPrueba)
print('\033[94m')

# instantiate tokenizer class
tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True,reduce_len=True)
# tokenize tweets
texto_tokens = tokenizer.tokenize(textoPrueba)

print()
print('Tokenized string:')
print(texto_tokens)
```

```python
#Import the english stop words list from NLTK
stopwords_english = stopwords.words('english') 

print('Stop words\n')
print(stopwords_english)

print('\nPunctuation\n')
print(string.punctuation)
```

```python
print()
print('\033[92m')
print(texto_tokens)
print('\033[94m')

texto_clean = []

for word in texto_tokens: # Go through every word in your tokens list
    if (word not in stopwords_english and  # remove stopwords
        word not in string.punctuation):  # remove punctuation
        texto_clean.append(word)

print('removed stop words and punctuation:')
print(texto_clean)
```

```python
print()
print('\033[92m')
print(texto_clean)
print('\033[94m')

# Instantiate stemming class
stemmer = PorterStemmer() 

# Create an empty list to store the stems
texto_stem = [] 


for word in texto_clean:
    stem_word = stemmer.stem(word)  # stemming word
    texto_stem.append(stem_word)  # append to the list

print('stemmed words:')
print(texto_stem)
```
